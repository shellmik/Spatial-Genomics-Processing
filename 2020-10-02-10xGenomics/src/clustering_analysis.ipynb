```python
import csv
import gzip
import os
import scipy.io
import pandas as pd
import numpy as np
```

```python
cluster_dir = "../data_processed/clustering/kmeans_{}_clusters/clusters.csv"
spatial_dir = "../data_processed/spatial_500.csv"
```

```python
df_spatial = pd.read_csv(spatial_dir)
```

```python
for i in range(2,11):
    df_cluster = pd.read_csv(cluster_dir.format(i))
    df_cluster = df_cluster.rename(columns={'Barcode': 'barcode', 'Cluster': 'cluster_{}'.format(str(i))})
    df_spatial = df_spatial.merge(df_cluster, how='inner', on='barcode')
```
